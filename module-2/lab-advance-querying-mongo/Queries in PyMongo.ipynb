```python
# https://api.mongodb.com/python/current/tutorial.html

from pymongo import MongoClient

client = MongoClient('mongodb://localhost:27017/')
db = client.companies
```

### 1. All the companies that it's name match 'Babelgum'. Retrieve only their `name` field.

```python
import pprint

comps = db.companies.find({"name":"Babelgum"},{'name':1})

for i,comps in enumerate(comps):
    pprint.pprint(comps)
    if i == 2:
        break
```

### 2. All the companies that have more than 5000 employees. Limit the search to 20 companies and sort them by **number of employees**.

```python
comps = db.companies.find({'number_of_employees':{'$gt':5000}}).sort('number_of_employees',1).limit(20)

for i,comps in enumerate(comps):
    pprint.pprint(comps)
    if i == 2:
        break
```

### 3. All the companies founded between 2000 and 2005, both years included. Retrieve only the `name` and `founded_year` fileds.

```python
comps = db.companies.find({ '$and': [ { 'founded_year': { '$gte': 2000 } }, { 'founded_year': { '$lte': 2005 } } ] },{'name':1, 'founded_year':1})

for i,comps in enumerate(comps):
    pprint.pprint(comps)
    if i == 3:
        break
```

### 4. All the companies that had a Valuation Amount of more than 100.000.000 and have been founded before 2010. Retrieve only the `name` and `ipo` fields.

```python
comps = db.companies.find({'$and': [{"ipo.valuation_amount":{'$gt':100000000}},{'founded_year':{'$lt':2010}}]},{'ipo':1,'name':1})

for i,comps in enumerate(comps):
    pprint.pprint(comps)
    if i == 2:
        break
```

### 5. All the companies that have less than 1000 employees and have been founded before 2005. Order them by the number of employees and limit the search to 10 companies.

```python
comps = db.companies.find({'$and': [{'number_of_employees':{'$lt':1000}},{'founded_year':{'$lt':2005}}]}).sort('number_of_employees',-1).limit(10)

for i,comps in enumerate(comps):
    pprint.pprint(comps)
    if i == 2:
        break
```

### 12. All the companies that have been 'deadpooled' after the third year.

```python
comps = db.companies.find({ '$where': "(this.deadpooled_year > this.founded_year+3) && (this.deadpooled_year >3) && (this.founded_year!= null) && (this.deadpooled_year!= null)"  })

for i,comps in enumerate(comps):
    pprint.pprint(comps)
    if i == 2:
        break
```

### 13. All the companies founded before 2000 that have and acquisition amount of more than 10.000.000

```python
comps = db.companies.find({'$and': [{'founded_year':{'$lt':2000}},{"acquisition.price_amount":{'$gt':10000000}}]})

for i,comps in enumerate(comps):
    pprint.pprint(comps)
    if i == 2:
        break
```
